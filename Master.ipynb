```python
class students: #define students class
                #define name, needs, availabilities as variables
    def __init__(self,name):
        self.name = name
        self.needs = [] 
        self.availabilities = []
    def need(self,classes): 
        self.needs.append(classes.lower())
    def available(self,time):
        self.availabilities.append(time.lower())
```

```python
class tutors: #define tutors class
              #define name, expertises, availabilities as vars
    def __init__(self,name):
        self.name = name
        self.expertises = []
        self.availabilities = []
    def expertise(self,classes):
        self.expertises.append(classes.lower())
    def availableDay(self,time):
        self.availabilities.append(time.lower())
```

```python
# function for adding students
def addStudent():
    studentDict = {}
    student = students(input("Enter your Full Name: "))
    numNeeds = int(input("How many classes do you need tutoring in?"))
    for j in range(numNeeds):
        needClass = input("Enter class"+' '+str((j+1))+':')
        student.need(needClass)
    
    numAvailabilities = int(input("How many times are you free for? "))
    for j in range(numAvailabilities):
        studentAvailable = input("What time are you available? (ex. Sunday Morning, Monday Evening, Thursday Afternoon etc.) ")
        student.available(studentAvailable)
    print("Thanks, you have been added.")
    studentDict[student] = [student.needs, student.availabilities]
    
    with open("Students", "a") as studentsFile:
        studentsFile.write(str(list(studentDict)[0].__dict__)+'\n')
```

```python
# function for adding tutors
tutorDict = {}
def addTutor():
    tutor = tutors(input("First and Last Name: "))
    numClasses = int(input("How many subjects are you comfortable tutoring? "))
    for i in range(numClasses):
        expertClass = input("Enter the subject to tutor: ")
        tutor.expertise(expertClass)
    numAvailabilities = int(input("How many times are you available to tutor for? "))
    for i in range(numAvailabilities):
        expertAvailable = input("What time are you available? (ex. Sunday Morning, Monday Afternoon, etc.) ")
        tutor.availableDay(expertAvailable)
    tutorDict[tutor] = [tutor.expertise,tutor.availableDay]
    print("Thanks, you have been added.")
    
    with open("Tutors", "a") as tutorsFile:
        tutorsFile.write(str(list(tutorDict)[0].__dict__)+'\n')
```
